#### 02 - Quais as disciplinas mais ofertadas da UFERSA?
* No gráfico abaixo são mostradas as 3 disciplinas mais ofertadas da UFERSA, são também as únicas ofertadas mais de 300 vezes. Isso indica uma grande demanda por essas disciplinas que pode estar ligada ao número médio de vezes que um aluno precisa cursá-la para conseguir aprovação.

```python
import matplotlib.pyplot as plt
import csv, seaborn as sb

x=[]
y=[]

with open('output/Q2.csv', 'r') as csvfile:
    plots = list(csv.reader(csvfile, delimiter=','))
    for i in plots[0:3]:
        y.append(i[1])
        x.append(int(i[0]))

plt.figure(figsize=(20,5))
plt.title('Dados: Disciplinas mais ofertadas')
plt.ylabel('Número de Disciplinas')

splot = sb.barplot(x, y)
for p in splot.patches:
    splot.annotate(format(p.get_width(), '.0f'), #, # Valores Numéricos
                       (p.get_x() + p.get_width(), p.get_y() + p.get_height() / 2), 
                       ha = 'center', va = 'center',
                       xytext = (10, 0), 
                       textcoords = 'offset points')
```
