#### 01 - Quantas disciplinas foram ofertadas pela Universidade desde o início do funcionamento (2007) até o período 2019.1?
Foram ofertadas um total de 25673 disciplinas desde 2007 até 2019. Apresentadas a seguir nas seguintes disposições, por ano e por período.

```python
#*-- coding utf8 --*

import matplotlib.pyplot as plt
import csv, seaborn as sb

x = []
y = []

with open('output/Q1_Ano.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.figure(figsize=(15,10))
plt.title('Dados: Disciplinas por Ano')
plt.xlabel('Ano')
plt.ylabel('Número de Disciplinas')

splot = sb.barplot(x, y)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., 
    p.get_height()), ha = 'center', va = 'center',
    xytext = (0, 10), textcoords = 'offset points')
plt.show()
```
